#### Contemplations: Using Lambda vs using local Flask app with REST API for deploying ML model:
Lambda Strengths:
- Lambda would be good use of AWS
- AWS is on a lot of job apps. 
- It would allow me to call the model from any computer because Lambda is hosted on web. 

Flask Strengths:
- There is no AWS prize
- It's always good to know the basics. If you know flask, maybe you know lambda. 
- More embressive. 



Using the flask server:
- Just use dill to dump your model into a file
- Load the file in server.py
- Make a json request, send it to your server.py location, and you're good.
